#3.2 Continuity and Differentiation

The limit of a function as x approaches a certain point is often written as L. What that means is that as x gets extremely close to that point, f(x) gets arbitrarily close to L (aka |f(x)- L | < epsilon)

A continuous function has no abrupt changes (aka discontinuities). A function is continuous at x=a if lim x->a f(x) = f(a). 

When you have a composite function (f(g(x)) for example) we say that the composition is continuous at x=a if both f and g are continuous at x=a.

Sometimes functions will also have a maximum or minimum. 

```python
from sympy import *
 
x = symbols('x')
expr1 = cos(x)/x;
expr2 = sin(x)/x;
expr3 = tan(x)/x;
   
print("Expression : {}".format(expr1))
     
limit_expr1 = limit(expr1, x, 0) 
     
print("Limit of the expression tends to 0 : {}".format(limit_expr1))

print("Expression : {}".format(expr2))
     
limit_expr2 = limit(expr2, x, 0) 
     
print("Limit of the expression tends to 0 : {}".format(limit_expr2))

print("Expression : {}".format(expr3))
     
limit_expr3 = limit(expr3, x, 0) 
     
print("Limit of the expression tends to 0 : {}".format(limit_expr3))
```

The derivative of a function measures the amount that the output changes with respect to another value (typically, how much y changes with respect to x aka dy/dx). 

We say that the derivative of a function at a given point: f'(a) can also be found using a limit definition: 

lim h->0 (f(a+h)-f(a))/h

Rolle's Theorem: Let f be a continuous function and assume its derivative exists on (a,b). If f(a) = f(b) then there exists some c s.t. a<c<b s.t. f'(c)=0

Mean Value Theorem: Let f be a continuous function and assume its derivative exists on (a,b). Then there exists some c s.t. f(b) = f(a) + (b-a)*f'(c)

You can also take partial derivatives when working with multivariable equations. 

**Let's look at some examples of derivative rules:**

```python
#Simple Functions

from sympy import *

x = Symbol('x')

f = x**2

derivative_f = f.diff(x)

derivative_f
```

```python
#Derivative at a given point

from sympy import *

x = Symbol('x')

f = x**2

derivative_f = f.diff(x)

f1 = lambdify(x, derivative_f)

f1(4)
```

```python
#Power Rule

import sympy as sym

x = sym.Symbol('x')
f = x**6
derivative_f = f.diff(x)
derivative_f
```

```python
#Product Rule
#(uv)' = u'v + uv'

import sympy as sym

x = sym.Symbol('x')
f = x**2*sym.sin(x)
derivative_f = f.diff(x)
derivative_f
```

```python
#Chain Rule

import sympy as sym

x = sym.Symbol('x')
f = sym.sqrt(x**3)
derivative_f = f.diff(x)
derivative_f
```

```python
#Partial Derivatives

import sympy as sym

x, y = sym.symbols('x y')
f = x**3+2*x*y**7

#partial x
derivative_f_x = f.diff(x)
derivative_f_x
```

```python
import sympy as sym

x, y = sym.symbols('x y')
f = x**3+2*x*y**7

#partial y
derivative_f_y = f.diff(y)
derivative_f_y
```

Taylor's Theorem: It gives an approximation of a differentiable function around a given point by a polynomial

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(2), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

#3.3 Unconstrained Optimization

Global Minimizer: Also known as the absolute minimum of a function. 

Local Minimizer: Also known as the local minimum of the function. 

Descent Direction & Directional Derivatives: If f is a continuously differentiable function at x=a then the vector v is a descent direction for f at x=a if the directional derivative of f at x=a in the direction of v is negative. 

A real valued function is convex if the line segment between any 2 points on the graph lies about the actual graph between those points. 

If f is a convex function then any local minimizer of f is also a global minimizer. 


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(0,8)
y = x**2

x1 = np.linspace(0,6)
y1 = x*4.5

plt.plot(x, y,'b',x1,y1,'r')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title("This is an example of a convex function")
#plt.legend()
plt.show()
```

Affine functions are convex. 

First-Order Convexity Condition: If f is a continuously differentiable function then f is convex if and only if for all x,y  R^d, f(y)  f(x) + f(x)^T (y-x).
