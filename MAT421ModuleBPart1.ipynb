Assume we independent data values xi and dependent data values yi, where i = 1, ..., n. We want to estimate a function that goes through all the data points. We don't need an underlying model but the model must inform the user of its quality (whether it's jagged or smooth for example). 

Here's a simple example:

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [6, 7, 9]

x_new = 2.5
y_new = np.interp(x_new, x, y)
print(y_new)
```

```python
plt.plot(x, y, 'o-')
plt.title('Interpolation at x = 2.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We also discuss linear and cubic spline interpolation. 

Linear: "the estimated point is assumed to lie on the line joining the nearest points to the left and right."

Here is an expample of linear interpolation used to find x=2.5 from the data x = [0,1,5] y = [1,3,9] :

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [0, 1, 5]
y = [1, 3, 9]

f = interp1d(x, y)
y_hat = f(2.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(2.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 2.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Cubic Spline: "the interpolation function is a set of piecewise cubic functions."

Here's an example of cubic spline interpolation using x = [1,2,4] y = [4,7,5] from 0<=x<=5

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 4]
y = [4, 7, 5]

f = CubicSpline(x, y, bc_type = 'natural')
x_new = np.linspace(0, 5, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
