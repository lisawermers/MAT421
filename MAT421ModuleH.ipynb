**22.1: ODE Initial Value Problem Statement**

A **differential equation** is a relationship between a function, it's independent variable, and some number of its derivatives. 

An **ordinary differential equation** is a differential equation in one dimension. 

A **partial differential equation** is a general form differential equation where x is a vector containing the independent variables, and the partial derivatives can be of any order with respect to any combination of variables. 

A **general solution** is some g(x) that satisfies the differential equation. For an ODE of order *n*, a **particular solution** is some p(x) that satisfies the differential equation and has *n* explcitely **known values** of the solution. 

Sometimes we'll be given an **initial value** along with an ODE. 

**22.2: Reduction of Order**

Many methods we have for solving IVP are made specifically for first order ODEs so to make the methods useful we often have to **reduce the order**. To do so, consider a vector *S(t)* which is the **state** of the system as a function of time. Then $\frac{dS(t)}{dt}$ can be written using only *S(t)* or its derivatives. 

Essentially, we turned an nth order ODE into n first order ODEs that are **coupled** together (meaning they share the same terms). 


**22.3: The Euler Method**

Let $\frac{dS(t)}{dt} = F(t,S(t))$ be an explicitely defined first order ODE. Also, let t be a numerical grid of the interval $[t_0,t_f]$ with spacing h. WLOG, let's assume $t_0 = 0$ and $t_f = N*h$ for some positive integer N. 

The linear approximation of S(t) around $t_j$ at $t_{j+1}$ is 

$S(t_{j+1}) = S(t_j) + (t_{j+1}-t_j)\frac{dS(t_j)}{dt}$

or

$S(t_{j+1}) = S(t_j) + hF(t_j,S(t_j))$

This formula is the **Explicit Euler Formula**. It allows us to compute an approximation for $S(t_{j+1})$ given by the state at $S(t_j)$. 

Let's use the differential equation $\frac{df(t)}{dt} = e^{-t}$ with initial condition $f_0 = -1$ and exact solution $f(t) = -e^{-t}$ and approximate the solution to this IVP between 0 and 1 using increments of 0.1:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-poster")
%matplotlib inline

#Define parameters
f = lambda t, s: np.exp(-t) #ODE
h = 0.1 #Step size
t = np.arange(0, 1+h, h) #Numerical grid
s0 = -1 #Initial condition

#Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
  s[i+1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12,8))
plt.plot(t, s, "b-", label="Approximate")
plt.plot(t, -np.exp(-t), "g", label="Exact")
plt.title("Approximate and Exact Solution for Simple ODE")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
